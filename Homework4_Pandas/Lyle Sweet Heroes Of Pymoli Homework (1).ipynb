```python
 # Average Purchase Total per Person by Gender

    #Items Purchased By Person
gender_purchase_count = purchase_data_df.groupby(["Gender", "SN"])
gender_IPP_purchase_count = gender_purchase_count.count()
gender_IPP_purchase_count_iso = pd.DataFrame(gender_IPP_purchase_count.iloc[:,0]).round(2)
gender_IPP_purchase_count_iso
    
    #Total Purchased By Person
gender_purchase_count = purchase_data_df.groupby(["Gender", "SN"])
gender_TPP_purchase_count = gender_purchase_count.sum()
gender_TPP_purchase_count_iso = pd.DataFrame(gender_TPP_purchase_count.iloc[:,3]).round(2)
gender_TPP_purchase_count_iso   

 # Average Purchase Total per Person by Gender
merge_table_df = pd.merge(gender_IPP_purchase_count_iso, gender_TPP_purchase_count_iso, on="SN", how="outer")

merge_table_df["Gender"] = gender_purchase_count["Gender"]

average_ppp_df = pd.DataFrame(gender_TPP_purchase_count_iso["Price"] / gender_IPP_purchase_count_iso["Purchase ID"]).round(2)

finished_table = pd.merge(merge_table_df, average_ppp_df, on="SN", how="outer")

finished_table = finished_table.rename(columns={"Purchase ID": "# of Purchases","Price":"Total Spend", 0:"Avg Purchase Total"})

finished_table
```

```python

### Top Spenders

# Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
  # SN
  # Purchase Count
  # Average Purchase Price
  # Total Purchase Value
```

```python
top_spenders_purchase_count = purchase_data_df.groupby("SN").count()["Price"].rename("Purchase Count")
top_spenders_average_price = purchase_data_df.groupby("SN").mean()["Price"].rename("Average Price")
top_spenders_total_purchase_value = purchase_data_df.groupby("SN").sum()["Price"].rename("Total Purchase Value")

top_spenders_df = pd.DataFrame({"Purchase Count": top_spenders_purchase_count, 
                                "Average Price": top_spenders_average_price, "Total Purchase Value":top_spenders_total_purchase_value})

top_spenders_df_sorted = top_spenders_df.sort_values("Total Purchase Value", ascending=False)
top_spenders_df_sorted.head()




```

```python
### Most Popular Items

# Identify the 5 most popular items by purchase count, then list (in a table):
  # Item ID
  # Item Name
  # Purchase Count
  # Item Price
  # Total Purchase Value
```

```python
most_popular_items_list = purchase_data_df.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
most_popular_items_price = purchase_data_df.groupby(["Item ID","Item Name"]).mean()["Price"].rename("Item Price")
most_popular_items_purchase_value = purchase_data_df.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Purchase Value")

most_popular_items_df = pd.DataFrame({"Purchase Count": most_popular_items_list,
                                      "Item Price": most_popular_items_price,
                                      "Total Purchase Value": most_popular_items_purchase_value})
                                                                 
most_popular_items_df_sorted = most_popular_items_df.sort_values("Purchase Count",ascending=False)
most_popular_items_df_sorted.head()
                                                                 
```

```python
### Most Profitable Items

# Identify the 5 most profitable items by total purchase value, then list (in a table):
  # Item ID
  # Item Name
  # Purchase Count
  # Item Price
  # Total Purchase Value
```

```python
most_profitable_items_purchase_count = purchase_data_df.groupby(["Item ID","Item Name"])["Price"].count().rename("Purchase Count")
most_profitable_items_item_price = purchase_data_df.groupby(["Item ID","Item Name"])["Price"].mean().rename("Item Price")
most_profitable_items_total_purchase_value = purchase_data_df.groupby(["Item ID","Item Name"])["Price"].sum().rename("Total Purchase Value")

most_profitable_items_df = pd.DataFrame({"Purchase Count": most_profitable_items_purchase_count, 
                             "Item Price": most_profitable_items_item_price,
                             "Total Purchase Value": most_profitable_items_total_purchase_value})

most_profitable_items_df_sorted = most_profitable_items_df.sort_values("Total Purchase Value", ascending=False)
most_profitable_items_df_sorted.head()


```
